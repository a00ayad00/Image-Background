```python
import matplotlib.pyplot as plt
import matplotlib.image as implt
import cv2
import numpy as np
import os
%matplotlib inline
# %matplotlib qt
os.chdir(r'D:\Projects\Computer Vision\Background Replacement')
```

```python
img = implt.imread('car_green_screen.jpg')
img.shape
```

```python
plt.imimshow(img)
plt.show()
```

```python
# Define the color threshold
lower_green = np.array([31, 0, 31])
upper_green = np.array([110, 255, 110])
```

```python
# Create a mask
mask = cv2.inRange(img, lower_green, upper_green)

plt.imshow(mask, cmap='gray')
plt.show()
```

```python
masked_image = np.copy(mask)
```

```python
mask.shape
```

```python
# Mask the image to let the car show through
masked_img = np.copy(img)

masked_img[mask != 0] = [0, 0, 0]

# Display it!
plt.imshow(masked_image)
plt.show()
```
